```python
import pickle
import pandas as pd

loaded_model = pickle.load(open("finalized_model_Logistic.sav", "rb"))

age = int(input("Enter age: "))
bp = int(input("Enter bp: "))
al = int(input("Enter al: "))
su = int(input("Enter su: "))
bgr = int(input("Enter bgr: "))
bu = int(input("Enter bu: "))
sc = float(input("Enter sc: "))
sod = int(input("Enter sod: "))
pot = float(input("Enter pot: "))
hrmo = float(input("Enter hrmo: "))
pcv = int(input("Enter pcv: "))
wc = int(input("Enter wc: "))
rc = float(input("Enter rc: "))

sg_b = int(input("Enter sg_b: "))
sg_c = int(input("Enter sg_c: "))
sg_d = int(input("Enter sg_d: "))
sg_e = int(input("Enter sg_e: "))

rbc_normal = int(input("Enter rbc_normal: "))
pc_normal = int(input("Enter pc_normal: "))
pcc_present = int(input("Enter pcc_present: "))
ba_present = int(input("Enter ba_present: "))
htn_yes = int(input("Enter htn_yes: "))
dm_yes = int(input("Enter dm_yes: "))
cad_yes = int(input("Enter cad_yes: "))
appet_yes = int(input("Enter appet_yes: "))
pe_yes = int(input("Enter pe_yes: "))
ane_yes = int(input("Enter ane_yes: "))

result = loaded_model.predict([[
    age, bp, al, su, bgr, bu, sc, sod, pot, hrmo, pcv,
    wc, rc, sg_b, sg_c, sg_d, sg_e, rbc_normal, pc_normal,
    pcc_present, ba_present, htn_yes, dm_yes, cad_yes,
    appet_yes, pe_yes, ane_yes
]])

if result[0] == 1:
    print("Yes")
else:
    print("No")

print("Prediction:", result)
```
