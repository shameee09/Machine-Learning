```python
import pandas as pd
data=pd.read_csv("Social_Network_Ads.csv")
data
```

```python
data=pd.get_dummies(data,drop_first=True).astype(int)
data
```

```python
independent=data[['Age', 'EstimatedSalary', 'Gender_Male']]
dependent=data[['Purchased']]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(independent,dependent,test_size=1/3,random_state=0)


from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x_train=sc.fit_transform(x_train)
x_test=sc.transform(x_test)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import GridSearchCV
param_grid={'criterion':['gini','entropy'],'max_features':['auto','sqrt','log2']}
grid=GridSearchCV(DecisionTreeClassifier(),param_grid,refit=True,verbose=3,n_jobs=-1)
grid.fit(x_train,y_train)
```

```python
table=pd.DataFrame.from_dict(grid.cv_results_)
table
```
