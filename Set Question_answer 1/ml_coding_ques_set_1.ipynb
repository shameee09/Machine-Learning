```python
#1.
import pandas as pd

# Creating a DataFrame
data = {'Customer_ID': [101, 102, 103, 101, 104, 102, 101, 105, 102, 103],
        'Purchase_Amount': [200, 150, 180, 220, 300, 200, 100, 400, 250, 300]}

df = pd.DataFrame(data)

purchases = df.groupby('Customer_ID')['Purchase_Amount'].sum().reset_index()

#  top 3 frequent customers
top_customers = purchases.sort_values(by='Purchase_Amount', ascending=False).head(3)

print("Total Purchases per Customer:")
print(purchases)
print("\nTop 3 Frequent Customers:")
print(top_customers)

```

```python

import pandas as pd
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import LabelEncoder

data = {'word_count': [100, 150, 200, 120, 180, 220],
        'is_spam': ['ham', 'spam', 'spam', 'ham', 'spam', 'spam']}
data= pd.DataFrame(data)

encoder= LabelEncoder()
data['is_spam'] = encoder.fit_transform(data['is_spam'])

X = data[['word_count']]
y = data['is_spam']

model = LogisticRegression()
model.fit(X, y)

words=model.predict([[200]])

if words==0:
  print("Prediction for email with 200 words: ham")
else:
  print("Prediction for email with 200 words: spam")





```
